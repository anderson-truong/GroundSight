```python
# Live filter from scratch
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from scipy import signal
from collections import deque
```

```python
%matplotlib inline

# Real IMU data
data_raw = []
with open('raw.txt', 'r') as f:
    data_raw = [line.rstrip('\n').split(',') for line in f.readlines()]
    
data_filtered = []
with open('filtered.txt', 'r') as f:
    data_filtered = [line.rstrip('\n').split(',') for line in f.readlines()]
    
data_diff = []
with open('diff.txt', 'r') as f:
    data_diff = [line.rstrip('\n').split(',') for line in f.readlines()]

data_raw_split = list(zip(*data_raw))
data_filtered_split = list(zip(*data_filtered))
data_diff_split = list(zip(*data_diff))

min_len = min((len(data_raw), len(data_filtered)))

ts = np.array(list(map(int, data_raw_split[0])))[:min_len]
y_raw = np.array(list(map(int, data_raw_split[1])))[:min_len]
y_filtered = np.array(list(map(int, data_filtered_split[1])))[:min_len]
y_diff = np.array(list(map(int, data_diff_split[1])))[:min_len]
plt.plot(ts, y_raw, label='Raw Acceleration')
plt.plot(ts, y_filtered, label='Filtered Acceleration')
plt.legend()
```

```python
plt.plot(range(len(y_raw)), y_raw)
```

```python
sl = slice(210, 600)
ts_slice = ts[sl]
y_raw_slice = y_raw[sl]
y_filtered_slice = y_filtered[sl]
#y_diff_slice = y_diff[sl]

plt.plot(ts_slice, y_raw_slice, label='Raw accel')
plt.plot(ts_slice, y_filtered_slice, '.', label='Filtered accel')
#plt.plot(ts_slice, y_diff_slice, '.', label='Diff')
plt.legend()
```

```python
# Difference
diffs = []
prev_diff = 0
for i in range(len(y_filtered_slice) - 1):
    difference = y_filtered_slice[i + 1] - y_filtered_slice[i]
    if difference >= 0 and prev_diff < 0:
        print(ts_slice[i] / 1.0e7)
    diffs.append(difference)
    prev_diff = difference
print(diffs)

plt.plot(ts_slice, y_raw_slice, label='Raw')
plt.plot(ts_slice, y_filtered_slice, label='Filtered')
plt.plot(ts_slice[:-1], diffs, '.', label='Slope')
plt.legend()
```
